```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=True, bins=50)
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips, kind='kde')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['tip'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, hue='sex', jitter=True)
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips)
```

```python
sns.factorplot(x='day',y='total_bill', data=tips, kind='bar')
```

```python
import seaborn as sns
%matplotlib inline

iris = sns.load_dataset('iris')
iris.tail()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
f =sns.PairGrid(iris)
f.map_diag(sns.distplot)
f.map_upper(plt.scatter)
f.map_lower(sns.kdeplot)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
#sns.set(rc = {'figure.figsize':(20,12)})
plt.figure(figsize=(20,9) )
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':25})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day', aspect=0.7, size=8)
```
